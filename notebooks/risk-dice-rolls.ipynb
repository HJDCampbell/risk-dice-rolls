```python
def calc_losses(attacker_die: list[int], defender_die: list[int]) -> list[int]:
    attacker_die_sorted = sorted(attacker_die, reverse=True)
    defender_die_sorted = sorted(defender_die, reverse=True)

    losses = [0, 0]
    for a_roll, d_roll in zip(attacker_die_sorted, defender_die_sorted):
        if a_roll > d_roll:
            # attacker wins
            losses[1] += 1
        else:
            # defender wins
            losses[0] += 1

    return losses

calc_losses([3, 4], [1, 5])
```
